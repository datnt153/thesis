```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
def draw_img(view):
    print(f"view: {view}")
    file_list = []
    legends = []
    for model_size in ['s', 'm', 'l']:
        for data in ['image', 'pose']:
            file_list.append(f"log_{view}_{data}-tf_efficientnetv2_{model_size}_in21k.csv")
            if data == "pose":
                legends.append(f"img + pose tf_efficientnetv2_{model_size}_in21k")
            else:
                legends.append(f"img tf_efficientnetv2_{model_size}_in21k")
    
    
#     file_list = glob.glob( f"*{view}*.csv")
#     legends = []

#     # Print the list of matching file names
#     for file in file_list:
#         legend = file.replace("log_", "").replace(".csv", "").replace("-", " ")
#         legends.append(legend)
#         print(legend)
#         print(file)
    
    dataframes = []

    for file in file_list:
        df = pd.read_csv(file)
        dataframes.append(df)
        
    plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

    for df in dataframes:
        plt.plot(df['epoch'], df['acc'])

    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'Accuracy vs. Epoch in {view} View')
    plt.legend(legends)  # Show the file names in the legend
    plt.grid(True)
    plt.savefig(f'acc_{view}.png')
    plt.show()
```

```python
views = ["Dashboard", "Rear_view", "Right_side_window"]

for view in views:
    draw_img(view)
```
