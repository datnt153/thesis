```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

def plot(f_name):
    print(f"file name: {f_name}")
    # Load the CSV file into a DataFrame
    df = pd.read_csv(f_name)

    # Assuming 'predict' and 'targets' are the columns containing prediction and target values
    predictions = df['predict']
    targets = df['targets']

    # Calculate F1 score with 'macro' average setting for multiclass
    f1 = f1_score(targets, predictions, average='macro')

    # Calculate accuracy
    accuracy = accuracy_score(targets, predictions)

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(targets, predictions)

    # Calculate percentages for each cell in the confusion matrix
    conf_matrix_percent = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis] * 100

    print(f'F1 Score (macro): {f1:.2f}')
    print(f'Accuracy: {accuracy}')

    # Visualize the confusion matrix as percentages
    plt.figure(figsize=(10, 10))
    sns.heatmap(conf_matrix_percent, annot=True, fmt='.2f', cmap='Blues')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title('Confusion Matrix (Percentages)')
    plt.show()
```

```python
import os 

for f_name in os.listdir("predicts/"):
    plot(f"predicts/{f_name}")
```
